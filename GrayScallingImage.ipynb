```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(name="Image",image = None,size =10):
  w,h = image.shape[0],image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(size * aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(name)
  plt.show()

#download image
!gdown --id 102uCujErifivK1ziRGssaQ09khI15g6q
!unzip --qq images.zip
```

```python
import requests

url = "https://th.bing.com/th?id=OIP.3mqSrPTyTbZo_wV3Qto6CgHaEK&w=333&h=187&c=8&rs=1&qlt=90&o=6&dpr=1.5&pid=3.1&rm=2"
response = requests.get(url)

if response.status_code == 200:
    with open("my_image.jpg", "wb") as file:
        file.write(response.content)
else:
    print("Không thể tải hình ảnh.")
```

```python
image = cv2.imread('/content/my_image.jpg')
imshow("image",image,size=5)
```

```python
gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
imshow("gray",gray_img,size=5)
```

```python
#tach mau RGB
R,G,B = cv2.split(image)
print(R)
print(G)
print(B)
print(R.shape)
print(G.shape)
print(B.shape)
imshow("R",R)
imshow("G",G)
imshow("B",B)
```

```python
import numpy as np
zeros = np.zeros(image.shape[:2],dtype = 'uint8')
imshow("R",cv2.merge([zeros,zeros,R]))
imshow("B",cv2.merge([B,zeros,zeros]))
imshow("G",cv2.merge([zeros,G,zeros]))
```

```python
merget = cv2.merge([R,G,B])
imshow("Merget",merget)
imshow("ORG_IMG",image)
```

```python
# Color space HSV
# Hue 0-179
# Saturation 0-255
#Value(Intersity) 0-255
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
imshow("HSV",hsv_image)
```

```python
#Drawing on images
zero_img = np.zeros((512,512,3),np.uint8)
zero_img_gray = cv2.cvtColor(zero_img,cv2.COLOR_BGR2GRAY)
imshow("zero_img",zero_img)
imshow("zero_img_gray",zero_img_gray)
```

```python
cv2.line(zero_img,(0,0),(511,511),(255,127,0),5)
imshow("line",zero_img)
```

```python

cv2.rectangle(zero_img,(100,100),(300,300),(200,100,4),5)
cv2.circle(zero_img,(150,150),100,(15,150,50),-1)
pts = np.array([[1,50],[400,50],[90,20],[50,500]],np.int32)
cv2.polylines(zero_img,[pts],True,(0,0,255),5)
imshow("rectangle",zero_img)
```
